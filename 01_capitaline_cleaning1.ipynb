```python
def process_one_company(file_path, company_name):
    df = pd.read_excel(file_path, sheet_name="Finance")

    # keep needed rows
    df_core = df[df.iloc[:, 0].isin([
        "Net Sales",
        "PBIDT",
        "Reported Profit After Tax",
        "EPS (Adj) (Unit Curr.)",
        "PBIDTM(%)",
        "PATM(%)",
        "Year"
    ])]

    # rename first column
    df_core = df_core.rename(columns={df_core.columns[0]: "metric"})

    # extract year row
    year_row = df_core[df_core["metric"] == "Year"]

    quarter_map = {}
    for col in year_row.columns[1:]:
        quarter_map[col] = int(year_row[col].values[0])

    # drop year row
    df_core = df_core[df_core["metric"] != "Year"]

    # melt
    df_long = df_core.melt(
        id_vars="metric",
        var_name="quarter_raw",
        value_name="value"
    )

    # map quarters
    df_long["quarter_code"] = df_long["quarter_raw"].map(quarter_map)

    def code_to_quarter(code):
        year = int(code // 100)
        month = int(code % 100)
        if month == 3:
            return f"{year}Q1"
        elif month == 6:
            return f"{year}Q2"
        elif month == 9:
            return f"{year}Q3"
        elif month == 12:
            return f"{year}Q4"
        else:
            return None

    df_long["quarter"] = df_long["quarter_code"].apply(code_to_quarter)

    # pivot
    df_final = df_long.pivot_table(
        index="quarter",
        columns="metric",
        values="value",
        aggfunc="first"
    ).reset_index()

    df_final["company"] = company_name

    return df_final
```

```python
df_test = process_one_company(
    RAW_DIR + "HUL.xlsx",
    "HUL"
)

df_test.head()
```

```python
import pandas as pd
import os

all_companies = []

for file in os.listdir(RAW_DIR):
    if file.endswith(".xlsx"):
        company = file.replace(".xlsx", "")
        df_company = process_one_company(RAW_DIR + file, company)
        all_companies.append(df_company)

fundamentals_all = pd.concat(all_companies, ignore_index=True)
fundamentals_all.head()
```
