```python
all_market = []

for company, ticker in ticker_map.items():
    price = yf.download(
        ticker,
        start="2014-01-01",
        end="2025-01-01",
        progress=False
    )

    # flatten columns
    price.columns = price.columns.get_level_values(0)

    # returns
    price["ret_1d"] = price["Close"].pct_change()
    price["ret_1m"] = price["Close"].pct_change(21)
    price["ret_3m"] = price["Close"].pct_change(63)

    # volatility
    price["vol_30d"] = price["ret_1d"].rolling(30).std()
    price["vol_60d"] = price["ret_1d"].rolling(60).std()

    # volume
    price["avg_vol_30d"] = price["Volume"].rolling(30).mean()
    price["avg_vol_5d"] = price["Volume"].rolling(5).mean()
    price["abnormal_volume"] = price["avg_vol_5d"] / price["avg_vol_30d"]

    # quarterly aggregation
    price_q = price.reset_index()
    price_q["quarter"] = price_q["Date"].dt.to_period("Q").astype(str)

    market_q = price_q.groupby("quarter").agg({
        "ret_1m": "last",
        "ret_3m": "last",
        "vol_30d": "last",
        "vol_60d": "last",
        "abnormal_volume": "last"
    }).reset_index()

    market_q["company"] = company
    all_market.append(market_q)

market_all = pd.concat(all_market, ignore_index=True)
market_all.head()
```

```python
final_dataset = fundamentals.merge(
    market_all,
    on=["company", "quarter"],
    how="left"
)

final_dataset.head()
```

```python
# extract quarter number (Q1, Q2, ...)
final_dataset["qtr"] = final_dataset["quarter"].str[-2:]

# create quarter dummies
final_dataset = pd.get_dummies(
    final_dataset,
    columns=["qtr"],
    drop_first=True
)

final_dataset.head()
```

```python
final_dataset["covid_dummy"] = (
    (final_dataset["quarter"] >= "2020Q1") &
    (final_dataset["quarter"] <= "2021Q2")
).astype(int)

final_dataset[["quarter", "covid_dummy"]].drop_duplicates().head(10)
```

```python
final_dataset_clean = final_dataset.dropna().reset_index(drop=True)
final_dataset_clean.shape
```

```python
final_dataset_clean.to_csv(
    "final_ml_dataset.csv",
    index=False
)
```
